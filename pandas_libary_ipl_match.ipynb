```python
#isin()
#find the most destructive death over batsman in the history og ipl
#strike rate = (number of runs/number of balls)/100
#min batsman 200 balls in the 16-20
```

```python
death_over = dy[dy['over']>15]
```

```python

all_batsman = death_over.groupby("batsman")["batsman_runs"].count()
all_batsman
```

```python
x = all_batsman>200
x
```

```python
#boolean series in actuall series
batsman_list = all_batsman[x].index.tolist()
batsman_list
```

```python
# runs scored by all these 43 batsman
# balls palyed by these 43 batsman
dy[dy["batsman"].isin(batsman_list)]
```

```python
final=dy[dy["batsman"].isin(batsman_list)]
```

```python
runs = final.groupby("batsman")["batsman_runs"].sum()
runs
```

```python
balls = final.groupby("batsman")["batsman_runs"].count()
balls
```

```python
sr = (runs/balls)*100
sr
```
